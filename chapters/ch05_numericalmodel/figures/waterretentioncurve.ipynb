# Water retention curve

```python
# Load all modules
%matplotlib inline

import numpy as np
import plotenv as pen

palette = pen.initialize(numColors=2)
```

```python
pc = -np.logspace(2,9)
theta_s = 0.6
theta_r = 0.2
alpha = 0.5
n = 2.5


theta = theta_r + (theta_s - theta_r)/(1+np.abs(alpha*pc)**n)**(1-1/n)

w = theta*1000.0

fig, ax = pen.figure('Net transpiration', figsize=(6,4))

#ax.plot(np.log10(-pc),theta/theta_s)
ax.plot(np.log10(-pc),w)
#ax.set(xlabel='$p_c$',ylabel='$w$')

#pen.cleanupFigure()
```
